# 动态条形图-改写
随机生成数据

```python
import random
import pandas as pd
import math

cats0 = ['驯鹿','飞机','火车','船','汽车']
color0 = ['#D6A5DD','#F7DC73','#EBBDBF','#76CBE8','#A4DFE5']
year0 = list(range(1977,2021))
year = []
for i in year0:
    for j in range(len(cats0)):
        year.append(i)
cats = cats0*len(year0)
color = color0*len(year0)
value = []
for i in year:
    value.append(math.pow(random.randint(0,50), 2))#求平方，拉开数据
df = pd.DataFrame()
df['year'] = year
df['value'] = value
df['cats'] = cats
df['color'] = color
df['csum']=df.groupby(['cats'])['value'].cumsum()
```

```python
import random
import pandas as pd
import math

cats0 = ['旭比','疯子','杰杰','潇姐','猪哥']
color0 = ['#D6A5DD','#F7DC73','#EBBDBF','#76CBE8','#A4DFE5']
year0 = list(range(1977,2021))
year = []
for i in year0:
    for j in range(len(cats0)):
        year.append(i)
cats = cats0*len(year0)
color = color0*len(year0)
value = []
for i in year:
    value.append(math.pow(random.randint(0,50), 2))#求平方，拉开数据
df = pd.DataFrame()
df['year'] = year
df['value'] = value
df['cats'] = cats
df['color'] = color
df['csum']=df.groupby(['cats'])['value'].cumsum()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline


symbols={'旭比':'image://旭比.png',
    '疯子':'image://疯子.png', 
         '杰杰':'image://杰杰.png', 
         '潇姐':'image://潇姐.png',
         '猪哥':'image://猪哥.png',
         
         }
tl = Timeline()
         
for i in range(1977,2019):
    df_sub = df[df['year']==i].sort_values(by='csum')
    cats_list = list(df_sub['cats'])
    csum_list = list(df_sub['csum'])
    color_list = list(df_sub['color'])
    y=[]
    for j in range(5):
        y.append(
            opts.BarItem(
                name=cats_list[j],
                value=csum_list[j],
                itemstyle_opts=opts.ItemStyleOpts(color=color_list[j]),
            
            )
        )
    bar = (
        Bar()
        .add_xaxis(cats_list)
        .add_yaxis("", y,label_opts=opts.LabelOpts(position="top",font_weight="bold"),category_gap=30,
                  markpoint_opts=opts.MarkPointOpts(#每个条形图的分类坐标内容value=cats_list[0], value=cats_list[1], value=cats_list[2], value=cats_list[3], value=cats_list[4]
            data=[opts.MarkPointItem(name="",coord=[csum_list[0], cats_list[0]],symbol=symbols[cats_list[0]],
                                     itemstyle_opts=opts.ItemStyleOpts(color=color_list[0],border_color='grey')),
                   opts.MarkPointItem(name="",coord=[csum_list[1], cats_list[1]],symbol=symbols[cats_list[1]],
                                      itemstyle_opts=opts.ItemStyleOpts(color=color_list[1],border_color='grey')),
                 opts.MarkPointItem(name="",coord=[csum_list[2], cats_list[2]],symbol=symbols[cats_list[2]],
                                    itemstyle_opts=opts.ItemStyleOpts(color=color_list[2],border_color='grey')),
                  opts.MarkPointItem(name="",coord=[csum_list[3], cats_list[3]],symbol=symbols[cats_list[3]],
                                     itemstyle_opts=opts.ItemStyleOpts(color=color_list[3],border_color='grey')),
                  opts.MarkPointItem(name="",coord=[csum_list[4], cats_list[4]],symbol=symbols[cats_list[4]],
                                     itemstyle_opts=opts.ItemStyleOpts(color=color_list[4],border_color='grey'))
                   ]))
            #category_gap=30,
        #.add_yaxis("", csum_list, label_opts=opts.LabelOpts(position="right"))
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts("能走多少米 (时间: {}天)".format(i-1976),pos_left=350,padding=[30,20]),
            xaxis_opts=opts.AxisOpts(max_=50000)
        )
    )
    tl.add(bar, "{}天".format(i-1976))
    tl.add_schema(play_interval=300, is_loop_play=True)#播放间隔，是否循环播放
tl.render_notebook()
```

```python
tl.render('群友步数PK.html')
```
